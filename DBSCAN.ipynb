```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

```python
def ReadData(datasetName):
    all_data=pd.read_csv(datasetName)
    print(all_data.shape)
    all_data.head()
    return all_data
```

```python
def trainDataset(all_data):
    train=all_data.iloc[:40000,1:3]
    print(train.shape)
    train.head()
    x=train.values
    plt.scatter(x[:,0],x[:,1])
    #plt.xlabel('Kill')
    #plt.ylabel('Walk')
    return train
```

```python
def dbscan(train):
  dbscan = DBSCAN(eps= 0.05, min_samples=100, metric='euclidean').fit(train)
  dbscan
  outliers_df=train[dbscan.labels_==-1]
  clusters_df=train[dbscan.labels_!=-1 ]
  colors=dbscan.labels_
  colors_clusters='white'
  color_outliers='red'
  fig=plt.figure()
  ax=fig.add_axes([.1, .1, 1, 1])
  #plt.xlabel('Kill')
  #plt.ylabel('Walk')
  plt.title('DBSCAN')
  ax.scatter(clusters_df.iloc[:,0], clusters_df.iloc[:,1] ,c =colors_clusters, edgecolors='black', s=50)
  ax.scatter(outliers_df.iloc[:,0], outliers_df.iloc[:,1],c =color_outliers, edgecolors='black', s=50)
  return colors
```

```python
all_data=ReadData('Kill_Walk.csv')
```

```python
train=trainDataset(all_data)
```

```python
colors=dbscan(train)
```

```python

print(f'Silhouette Score: {silhouette_score(train, colors)}')
```
